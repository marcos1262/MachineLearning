```python
from sklearn.datasets import load_iris
base = load_iris()
```

```python
import numpy as np

y = base.target

M, N = base.data.shape
t = np.zeros(N)

# Parâmetros
alpha = 0.01
absoluteError = 0.1


# Função logística
def sigmoid(x):
    global t
    return 1 / (1 + np.exp(-np.dot(x, t)))


# Loop de treinamento
while True:
    t_aux = t.copy()
    grad = np.zeros(N)  # inicia o gradiente

    # atualiza o gradiente com informação de todos os pontos
    for j in range(N):
        grad[j] += np.dot((sigmoid(base.data) - y), base.data[:, j])

    # atualiza os parâmetros
    t_aux -= alpha * grad
    
    print(t)
    print(t_aux)

    diff = 0
    for j in range(N):
        diff += np.abs(t_aux[j] - t[j])
        print(diff)
        t[j] = t_aux[j]

    if diff < absoluteError:
        break

    print(diff)
    

print("Resultado:", t)
```
