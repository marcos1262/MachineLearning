```python
from sklearn.datasets import load_iris
base = load_iris()
```

```python
import numpy as np

y = base.target
M, N = base.data.shape
t = np.zeros(N)
alpha = 0.1
absoluteError = 0.1


def h_t(x):
    return 1 / (1 + np.e ** (-np.sum(t * x)))


for k in range(1000):
    t_aux = t.copy()
    for j in range(N):
        total = np.sum([
            (h_t(base.data[i]) - y[i]) * base.data[i][j]
            for i in range(M)])
        t_aux[j] = t[j] - alpha * total
    
    print(t)
    print(t_aux)

    diff = 0
    for j in range(N):
        diff += np.abs(t_aux[j] - t[j])
        print(diff)
        t[j] = t_aux[j]

    if diff < absoluteError:
        break

    print(diff)
    

print("Resultado:", t)
```
